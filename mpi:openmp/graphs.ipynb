## Ускорение $S = \dfrac{T_1}{T_p}$, эффективность $E = \dfrac{S}{p}$

# Задача 2

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

p = np.arange(1, 12)

line_time2 = 0.013
line_time10 = 0.062
line_time50 = 0.393

time2 = np.array([0.005754077999999999, 0.004038089, 0.004588408, 0.002723974, 0.003751627, 0.003724039, 0.0024586869999999998,0.002237334, 0.002103728,0.002164987, 0.002085182])
time10 = np.array([0.0345826, 0.0202289, 0.0192272, 0.0116127, 0.0133193, 0.012019, 0.00978929, 0.00743874, 0.00797676, 0.00642588, 0.00641497])
time50 = np.array([0.265606, 0.1776709, 0.656623, 0.09747496, 0.10375893, 0.7366964, 0.05244316, 0.04662185, 0.5628436, 0.02733192, 0.02516349, ])


fig, (ax1, ax2) = plt.subplots(1, 2,  figsize=[18, 10])
ax1.grid(ls=':')
ax2.grid(ls=':')

fig.suptitle('Графики эффективности и ускорнеия')

ax1.plot(p, line_time2 / time2, label = r'2000 точек')
ax1.plot(p, line_time10 / time10, label = r'10000 точек')
ax1.plot(p, line_time50 / time50, label = r'50000 точек')
ax1.legend()

ax2.plot(p, (line_time2 / time2) / p, label = r'2000 точек')
ax2.plot(p, line_time10 / time10 / p, label = r'10000 точек')
ax2.plot(p, line_time50 / time50 / p, label = r'50000 точек')

ax2.legend()
```
